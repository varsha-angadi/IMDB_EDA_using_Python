## 2. This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?

```python
df1['roi'] = np.where(df1['budget'] == 0, np.nan, ((df1['revenue'] - df1['budget']) / df1['budget']) * 100)

# Exclude rows with inf ROI
df1 = df1.replace([np.inf, -np.inf], np.nan)

# Calculate mean ROI for each language, excluding NaN values
avg_roi_by_language = df1.groupby('original_language')['roi'].mean().dropna()

# Find the language with the highest average ROI
max_avg_roi_language = avg_roi_by_language.idxmax()
print(f"The language with the highest average ROI is {max_avg_roi_language}")
```

## 4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
import pandas as pd
from collections import Counter

# Assuming df is your DataFrame containing the dataset
# Define the function to extract producers and directors
def extract_producers(crew_data):
    producers = []
    if isinstance(crew_data, list):  # Only process if crew_data is a list
        for member in crew_data:
            if member['job'] == 'Producer':
                producers.append(member['name'])
    return producers

# Extract producers and create a new column in the DataFrame
df['producers'] = df['crew'].apply(extract_producers)

# Flatten the list of producers
all_producers = [producer for sublist in df['producers'] for producer in sublist]

# Count the occurrences of each producer
producer_counts = Counter(all_producers)

# Get the top three producers
top_three_producers = producer_counts.most_common(3)

# Display the top three producers
print(top_three_producers)
```

## 5.Which actor has acted in the most number of movies?

```python
import pandas as pd
import ast

# Safe parsing function
def safe_parse_list(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            return []
    return []

# Function to extract actor names
def extract_actor_names(cast_list):
    if isinstance(cast_list, list):
        return [actor['name'] for actor in cast_list if isinstance(actor, dict) and 'name' in actor]
    return []

# Load the data
df = pd.read_csv("C:\\Users\\varsh\\Downloads\\imdb_data.csv")

# Apply safe parsing
df['cast'] = df['cast'].apply(safe_parse_list)

# Extract actor names
df['actors'] = df['cast'].apply(extract_actor_names)

# Explode the 'actors' column and count appearances
actor_counts = df.explode('actors')['actors'].value_counts()

# Find the actor with the most movies
most_movies_actor = actor_counts.idxmax()
most_movies_count = actor_counts.max()

print(f"Actor with the most movies: {most_movies_actor}")
print(f"Number of movies: {most_movies_count}")
```
